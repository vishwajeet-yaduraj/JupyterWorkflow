#  Bar charts using matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from collections import Counter
```

```python
data = pd.read_csv('data.csv')
```

```python
lang_count = Counter()
languages = data['LanguagesWorkedWith']

for row in languages:
    lang_count.update(row.split(';'))

lang_count.most_common(15)
```

```python
language=[]
popularity=[]

for item in lang_count.most_common(15):
    language.append(item[0])
    popularity.append(item[1])
language.reverse()
popularity.reverse()
```

```python
plt.barh(language,popularity)
plt.title('Most Popular Languages')
plt.xlabel('Number of People who use')
plt.tight_layout()
plt.show()
```
