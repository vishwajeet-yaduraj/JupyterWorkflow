```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x,y,c = colors, s=sizes, edgecolors='black', cmap='Greens',
            linewidths=1, alpha = 0.75)
cbar = plt.colorbar()
cbar.set_label('Color level')
```

```python
data = pd.read_csv('data6.csv')
```

```python
views = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(views, likes,c=ratio,cmap='summer', edgecolors='black',linewidths=1, alpha = 0.75)
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('LIKE DISLIKE RATIO')

plt.title('YouTube Trending videos')
plt.xlabel('No. of Views')
plt.ylabel('Likes')
```
