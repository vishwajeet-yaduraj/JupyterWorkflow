Fill Between method in Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data (1).csv')
```

```python
ages = data['Age']
dev_sal = data['All_Devs']
py_sal = data['Python']
js_sal = data['JavaScript']
overall_median = 57287
```

```python
plt.plot(ages,dev_sal,color = '#444444', linestyle = '--',label = 'All Devs')
plt.plot(ages,py_sal, label = 'Py Devs')
# plt.plot(age,js_sal, label = 'JS Devs')

plt.legend()
plt.fill_between(ages,py_sal,overall_median,
                 where =(py_sal>overall_median), interpolate=True,alpha = 0.2)
plt.fill_between(ages,py_sal,overall_median,
                 where =(py_sal<overall_median), interpolate=True,alpha = 0.2)

plt.title('Median Salary by age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
```

```python
plt.plot(ages,dev_sal,color = '#444444', linestyle = '--',label = 'All Devs')
plt.plot(ages,py_sal, label = 'Py Devs')
# plt.plot(age,js_sal, label = 'JS Devs')

plt.fill_between(ages,py_sal,dev_sal,where = (py_sal>dev_sal),interpolate=True,alpha = 0.2,label = 'Above Avg')
plt.fill_between(ages,py_sal,dev_sal,where = (py_sal<=dev_sal),interpolate=True,alpha = 0.2,label = 'Below Avg')

plt.legend()

plt.title('Median Salary by age')
plt.xlabel('Age')
plt.ylabel('Median Salary')
```
